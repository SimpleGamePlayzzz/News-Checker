```python
import requests
from bs4 import BeautifulSoup
import csv
import logging
import datetime

logging.basicConfig(filename="scraper.log", level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

BBC_NEWS_URL = "https://www.bbc.com/news"

def fetch_headlines(url, limit=10):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, "html.parser")

        headline_tags = soup.find_all("h3", class_="sc-759qgu-0")
        headlines = [tag.get_text(strip=True) for tag in headline_tags][:limit]

        if not headlines:
            logging.warning(f"No headlines found on {url}")
            return ["No headlines found."]
        
        return headlines
    
    except requests.exceptions.RequestException as e:
        logging.error(f"Error fetching {url}: {e}")
        return [f"Error: {e}"]

def save_to_csv(data, filename="news_headlines.csv"):
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["No.", "Headline"])
        for i, headline in enumerate(data, start=1):
            writer.writerow([i, headline])
    logging.info(f"Headlines saved to {filename}")

def main():
    headlines = fetch_headlines(BBC_NEWS_URL)

    print("\nLatest BBC News Headlines:\n")
    for i, headline in enumerate(headlines, start=1):
        print(f"{i}. {headline}")

    save_to_csv(headlines)

if __name__ == "__main__":
    start_time = datetime.datetime.now()
    logging.info("News scraper started.")
    
    main()
    
    end_time = datetime.datetime.now()
    logging.info(f"News scraper finished. Execution time: {end_time - start_time}")
```
